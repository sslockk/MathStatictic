4. Задание (в программе)
Решите систему уравнений:

```python
# Решите систему уравнений:y = x2 – 1    exp(x) + x∙(1 – y) = 1
x = np.linspace(-np.pi, np.pi, 201)
y = np.linspace(-np.pi, np.pi, 201)
f, ax = plt.subplots(figsize=(8, 8))

yx1 = x ** 2 - y -1
yx2 = np.exp(x) + x * (1 - y) - 1
plt.plot(x, yx1);
plt.plot(x, yx2);
ax.vlines(0, -15, 15, color='r', linestyle  = '--');
ax.vlines(-0.5, -15, 15, color='r', linestyle  = '--');
# на этом графике видно, что решение системы уравнений находиться в интервале от -0.5 до 0
```

```python
from sympy import Symbol, nsolve, exp
x = Symbol('x')
y = Symbol('y')
yx1 = x ** 2 - y -1
yx2 = exp(x) + x * (1 - y) - 1

nsolve((yx1, yx2), (x, y), (-0.5, 0))
```

```python
# Решите систему уравнений: y = x2 – 1     exp(x) + x∙(1 – y) > 1

x = np.linspace(-np.pi, np.pi, 201)
y = np.linspace(-np.pi, np.pi, 201)
f, ax = plt.subplots(figsize=(8, 8))

yx1 = x ** 2 - y -1
yx2 = np.exp(x) + x * (1 - y) - 1
plt.plot(x, yx1);
plt.plot(x, yx2);
ax.vlines(0, -15, 15, color='r', linestyle  = '--');
ax.vlines(1, -15, 15, color='r', linestyle  = '--');
# на этом графике видно, что решение системы уравнений находиться в интервале от -0.5 до 0
```

1) Напишите код, проверяющий любую из теорем сложения или умножения вероятности на примере подбрасывания монетки.

```python
def orel_reshka(n):
    m = 0
    for i in range(n):
        x = random.uniform(0, 1)
        if x < 0.5:
            m += 1
    return m / n
```

```python
x = 100
summ = []
mult = []
for i in range(x):
    n = 1000
    orel = orel_reshka(n)
    reshka = orel_reshka(n)
    summ.append(orel + reshka)
    mult.append(orel * reshka)
```

```python
fig = plt.figure(figsize=(4,4))
plt.boxplot(summ);
plt.title('Сумма вероятностей подбрасывания монетки')
# plt.boxplot(mult);
```

```python
fig = plt.figure(figsize=(4,4))
plt.boxplot(mult);
plt.title('Умножение вероятностей подбрасывания монетки')
```

```python
# Чем больше испытания, тем сильнее схождение медианы распределения к формуле сложения и умножения вероятности
# В нашем случае сложение 2-х независымых событий выпадения орла и решки равно 1
# А умножение вероятностей равно 0.25
```

4. Повторите расчеты, взяв другие значения n и k.

```python
k = 3
n = 10
str_n = (''.join(str(x) for x in range(n)))
permult = []
for p in itertools.permutations(str_n, k):
    permult.append(p)
#     print(''.join(str(x) for x in p))
print(f'n = {n}, k = {k}, кол-во сочетаний с перестановкой', len(permult))

combin = []
for p in itertools.combinations(str_n, k):
    combin.append(p)
#     print(''.join(str(x) for x in p))
print(f'n = {n}, k = {k}, кол-во сочетаний с перестановкой', len(combin))
```

5. Дополните код расчетом коэффициента корреляции x и y по формуле

```python
n = 100
r = 0.7
x = np.random.rand(n)
y = r * x + (1 - r) * np.random.rand(n)
plt.plot(x, y, 'o');
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]
print(a1, b1)
plt.plot([0, 1], [b1, a1 + b1]);
```

```python
x_med = np.mean(x)
y_med = np.mean(y)
corr_coef = (np.sum((x - x_med) * (y - y_med))) / np.sqrt(np.sum((x - x_med) ** 2) * np.sum((y - y_med) ** 2))
print('Коэффициент корреляции равен =', corr_coef)
```
